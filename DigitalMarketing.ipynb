```python
def getDay(day):
    return pd.to_datetime(day).strftime("%d %b (%a)")

def getCostPerLead(amountSpent, results):
    return amountSpent/results

def getMemerPassed(landingPageViews, linkClicks):
    return round((landingPageViews/linkClicks)*100,2)

def getLPConversion(results, landingPageViews):
    return round((results/landingPageViews)*100,2)
```

```python
# Create an empty DataFrame with column names and data types
schema={'Date': 'object', 'Campaign Name': 'object', 'AD Spent': 'float64','No.of Leads': 'int64',
        'Cost Per Lead': 'float64','CTR': 'float64','Link Click': 'int64','LP Views': 'int64',
        'Member Passed': 'float64','LP Conversion %': 'float64'}
output_df = pd.DataFrame(columns=schema.keys()).astype(schema)
```

```python
from datetime import datetime

dict_list = []

# loop through the rows using iterrows()
for index, row in dataset.iterrows():
    row_dict = {'Date': getDay(row['Day']), 'Campaign Name': row['Campaign name'], 
                'AD Spent': row['Amount spent (INR)'],'No.of Leads': row['Results'],
                'Cost Per Lead': getCostPerLead(row['Amount spent (INR)'],row['Results']),
                'CTR ':  round(row['CTR (link click-through rate)'],2),
                'Link Click': row['Link clicks'],'LP Views': row['Landing page views'],
                'Member Passed':  getMemerPassed(row['Landing page views'], row['Link clicks']),
                'LP Conversion %':  getLPConversion(row['Results'],row['Landing page views'])}
    dict_list.append(row_dict)


output_df = pd.DataFrame.from_dict(dict_list)
print(output_df)
```
